# Seaborn

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# relational plot
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="sex")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="sex", style="smoker")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="sex", style="smoker", size='size')
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue='time', style="smoker", col='sex')
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue='time', style="smoker", col='day')
```

```python
sns.scatterplot(x='day', y='total_bill', data=tips)
```

```python
sns.scatterplot(x='day', y='tip', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.catplot(x='total_bill', y='tip', data=tips)
```

```python
sns.catplot(x='day', y='tip', data=tips)
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='swarm')
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='box', hue='size')
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='box', hue='smoker')
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='box', hue='sex')
```

```python
with sns.axes_style(style="ticks"):
    g = sns.catplot(x="day", y="total_bill", hue="sex", data=tips, kind="box")
    g.set_axis_labels("Day", "Total Bill")
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='violin', hue='sex')
```

```python
sns.jointplot("total_bill", "tip", data=tips)
```

```python
sns.axes_style("white")
sns.jointplot("total_bill", "tip", data=tips, kind='hex')
```

```python
sns.axes_style("white")
sns.jointplot("total_bill", "tip", data=tips, kind='reg')
```
